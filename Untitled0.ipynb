```python
# === WEB ARAYÜZLÜ CHATBOT (GROQ İLE) ===
# !pip install -q langchain langchain-community langchain-groq wikipedia faiss-cpu sentence-transformers gradio

import os
import gradio as gr
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
import wikipedia

# === GROQ API ANAHTARI ===
if "GROQ_API_KEY" not in os.environ:
    print("⚠️  GROQ API KEY YOK!")
    print('os.environ["GROQ_API_KEY"] = "sizin_api_keyiniz"')
    print("\nAPI key almak için: https://console.groq.com/playground")
    raise Exception("API Key gerekli!")

print("🔧 Sistem hazırlanıyor...\n")

# Türkçe Wikipedia
wikipedia.set_lang("tr")

# --- MODEL (Groq Llama3) ---
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  
    temperature=0.7
)

# HuggingFace Embeddings 
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

print("📥 Wikipedia'dan veri çekiliyor...\n")

# --- VERİ TOPLAMA ---
cities = [
    # Batı Avrupa
    "Londra", "Paris", "Brüksel", "Amsterdam", "Lüksemburg", "Monako",
    # Kuzey Avrupa
    "Oslo", "Stockholm", "Kopenhag", "Helsinki", "Reykjavik", "Tallinn", "Riga", "Vilnius",
    # Güney Avrupa
    "Lizbon", "Madrid", "Roma", "Atina", "Valletta", "San Marino", "Andorra la Vella", "Podgorica", "Zagreb", "Sarajevo", "Belgrad", "Tiran",
    # Orta Avrupa
    "Viyana", "Berlin", "Bern", "Prag", "Bratislava", "Budapeşte", "Ljubljana",
    # Doğu Avrupa
    "Minsk", "Moskova", "Kiev", "Chişinău", "Sofya", "Bükreş", "Varsova",
    # Büyük şehirler (başkent olmayan ama önemli)
    "Barselona", "Milan", "Barcelona", "Valencia", "Munih", "Hamburg", "Manchester"
]


documents = []
for city in cities:
    try:
        print(f"  • {city}...", end=" ")
        summary = wikipedia.summary(city, sentences=5)
        page = wikipedia.page(city)

        content = f"""
ŞEHİR: {city}

ÖZET: {summary}

DETAY: {page.content[:1500]}
"""
        documents.append(Document(page_content=content, metadata={'city': city}))
        print("✓")
    except:
        print("✗")
        continue

print(f"\n✓ {len(documents)} şehir yüklendi\n")

# --- VEKTÖR VERİTABANI ---
print("🔮 Vektör veritabanı oluşturuluyor...")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=150)
texts = text_splitter.split_documents(documents)
vector_store = FAISS.from_documents(texts, embeddings)
print("✓ Hazır!\n")

# --- SORU-CEVAP ZİNCİRİ ---
prompt_template = """
Verilen bilgileri kullanarak soruya samimi ve detaylı cevap ver.
Bir tur rehberi gibi Türkçe konuş.
Bilgi yoksa "Bu konuda bilgim yok" de.

Bilgiler:
{context}

Soru: {question}

Cevap:
"""
prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
chain = load_qa_chain(llm, chain_type="stuff", prompt=prompt)

# --- CHATBOT FONKSİYONU ---
def chatbot_response(user_message, history):
    """Kullanıcı mesajına cevap üret"""
    try:
        # İlgili dökümanları bul
        docs = vector_store.similarity_search(user_message, k=3)

        # Cevap üret
        response = chain.invoke({
            "input_documents": docs,
            "question": user_message
        })

        return response["output_text"]

    except Exception as e:
        return f"❌ Hata: {str(e)}"

# --- GRADIO ARAYÜZÜ ---
print("="*60)
print("🚀 WEB ARAYÜZÜ BAŞLATILIYOR (GROQ)")
print("="*60)

demo = gr.ChatInterface(
    fn=chatbot_response,
    title="🌍 Avrupa Tur Rehberi Chatbot",
    description="""
    Merhaba! Ben Avrupa şehirleri hakkında bilgi veren bir tur rehberiyim.

    📍 Bilgi verdiğim şehirler: Londra, Paris, Brüksel, Amsterdam, Lüksemburg, Monako, Oslo, Stockholm, Kopenhag, Helsinki, Reykjavik, Tallinn, Riga, Vilnius, Lizbon, Madrid, Roma, Atina, Valletta, San Marino, Andorra la Vella, Podgorica, Zagreb, Sarajevo, Belgrad, Tiran, Viyana, Berlin, Bern, Prag, Bratislava, Budapeşte, Ljubljana, Minsk, Moskova, Kiev, Chişinău, Sofya, Bükreş, Varsova, Barselona, Milan, Barcelona, Valencia, Munih, Hamburg, Manchester

    💬 Örnek sorular:
    - "Roma'da ne yapmalıyım?"
    - "Paris'te gezilecek yerler neler?"
    - "Barselona'da hangi yemekleri denemeli?"

    🤖 Groq llama-3.3-70b-versatile modeliyle çalışıyor (OpenAI alternatifi)
    """,
    examples=[
        "Roma'da ne yapmalıyım?",
        "Paris'te gezilecek yerler neler?",
        "Barselona'da hangi yemekleri denemeli?",
        "Amsterdam'da bisikletle gezebilir miyim?",
        "Viyana'da klasik müzik konseri nerede izlenir?"
    ]
)

# Arayüzü başlat
demo.launch(share=True, debug=False)
```
