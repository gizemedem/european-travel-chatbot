```python
# === WEB ARAY√úZL√ú CHATBOT (GROQ ƒ∞LE) ===
# !pip install -q langchain langchain-community langchain-groq wikipedia faiss-cpu sentence-transformers gradio

import os
import gradio as gr
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
import wikipedia

# === GROQ API ANAHTARI ===
if "GROQ_API_KEY" not in os.environ:
    print("‚ö†Ô∏è  GROQ API KEY YOK!")
    print('os.environ["GROQ_API_KEY"] = "sizin_api_keyiniz"')
    print("\nAPI key almak i√ßin: https://console.groq.com/playground")
    raise Exception("API Key gerekli!")

print("üîß Sistem hazƒ±rlanƒ±yor...\n")

# T√ºrk√ße Wikipedia
wikipedia.set_lang("tr")

# --- MODEL (Groq Llama3) ---
llm = ChatGroq(
    model="llama-3.3-70b-versatile",  
    temperature=0.7
)

# HuggingFace Embeddings 
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

print("üì• Wikipedia'dan veri √ßekiliyor...\n")

# --- VERƒ∞ TOPLAMA ---
cities = [
    # Batƒ± Avrupa
    "Londra", "Paris", "Br√ºksel", "Amsterdam", "L√ºksemburg", "Monako",
    # Kuzey Avrupa
    "Oslo", "Stockholm", "Kopenhag", "Helsinki", "Reykjavik", "Tallinn", "Riga", "Vilnius",
    # G√ºney Avrupa
    "Lizbon", "Madrid", "Roma", "Atina", "Valletta", "San Marino", "Andorra la Vella", "Podgorica", "Zagreb", "Sarajevo", "Belgrad", "Tiran",
    # Orta Avrupa
    "Viyana", "Berlin", "Bern", "Prag", "Bratislava", "Budape≈üte", "Ljubljana",
    # Doƒüu Avrupa
    "Minsk", "Moskova", "Kiev", "Chi≈üinƒÉu", "Sofya", "B√ºkre≈ü", "Varsova",
    # B√ºy√ºk ≈üehirler (ba≈ükent olmayan ama √∂nemli)
    "Barselona", "Milan", "Barcelona", "Valencia", "Munih", "Hamburg", "Manchester"
]


documents = []
for city in cities:
    try:
        print(f"  ‚Ä¢ {city}...", end=" ")
        summary = wikipedia.summary(city, sentences=5)
        page = wikipedia.page(city)

        content = f"""
≈ûEHƒ∞R: {city}

√ñZET: {summary}

DETAY: {page.content[:1500]}
"""
        documents.append(Document(page_content=content, metadata={'city': city}))
        print("‚úì")
    except:
        print("‚úó")
        continue

print(f"\n‚úì {len(documents)} ≈üehir y√ºklendi\n")

# --- VEKT√ñR VERƒ∞TABANI ---
print("üîÆ Vekt√∂r veritabanƒ± olu≈üturuluyor...")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=150)
texts = text_splitter.split_documents(documents)
vector_store = FAISS.from_documents(texts, embeddings)
print("‚úì Hazƒ±r!\n")

# --- SORU-CEVAP Zƒ∞NCƒ∞Rƒ∞ ---
prompt_template = """
Verilen bilgileri kullanarak soruya samimi ve detaylƒ± cevap ver.
Bir tur rehberi gibi T√ºrk√ße konu≈ü.
Bilgi yoksa "Bu konuda bilgim yok" de.

Bilgiler:
{context}

Soru: {question}

Cevap:
"""
prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
chain = load_qa_chain(llm, chain_type="stuff", prompt=prompt)

# --- CHATBOT FONKSƒ∞YONU ---
def chatbot_response(user_message, history):
    """Kullanƒ±cƒ± mesajƒ±na cevap √ºret"""
    try:
        # ƒ∞lgili d√∂k√ºmanlarƒ± bul
        docs = vector_store.similarity_search(user_message, k=3)

        # Cevap √ºret
        response = chain.invoke({
            "input_documents": docs,
            "question": user_message
        })

        return response["output_text"]

    except Exception as e:
        return f"‚ùå Hata: {str(e)}"

# --- GRADIO ARAY√úZ√ú ---
print("="*60)
print("üöÄ WEB ARAY√úZ√ú BA≈ûLATILIYOR (GROQ)")
print("="*60)

demo = gr.ChatInterface(
    fn=chatbot_response,
    title="üåç Avrupa Tur Rehberi Chatbot",
    description="""
    Merhaba! Ben Avrupa ≈üehirleri hakkƒ±nda bilgi veren bir tur rehberiyim.

    üìç Bilgi verdiƒüim ≈üehirler: Londra, Paris, Br√ºksel, Amsterdam, L√ºksemburg, Monako, Oslo, Stockholm, Kopenhag, Helsinki, Reykjavik, Tallinn, Riga, Vilnius, Lizbon, Madrid, Roma, Atina, Valletta, San Marino, Andorra la Vella, Podgorica, Zagreb, Sarajevo, Belgrad, Tiran, Viyana, Berlin, Bern, Prag, Bratislava, Budape≈üte, Ljubljana, Minsk, Moskova, Kiev, Chi≈üinƒÉu, Sofya, B√ºkre≈ü, Varsova, Barselona, Milan, Barcelona, Valencia, Munih, Hamburg, Manchester

    üí¨ √ñrnek sorular:
    - "Roma'da ne yapmalƒ±yƒ±m?"
    - "Paris'te gezilecek yerler neler?"
    - "Barselona'da hangi yemekleri denemeli?"

    ü§ñ Groq llama-3.3-70b-versatile modeliyle √ßalƒ±≈üƒ±yor (OpenAI alternatifi)
    """,
    examples=[
        "Roma'da ne yapmalƒ±yƒ±m?",
        "Paris'te gezilecek yerler neler?",
        "Barselona'da hangi yemekleri denemeli?",
        "Amsterdam'da bisikletle gezebilir miyim?",
        "Viyana'da klasik m√ºzik konseri nerede izlenir?"
    ]
)

# Aray√ºz√º ba≈ülat
demo.launch(share=True, debug=False)
```
